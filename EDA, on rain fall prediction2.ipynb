```python
df1=df.copy()
df1.drop(['Evaporation','Sunshine','Cloud9am','Cloud3pm','Temp3pm'], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(13,5))
plt.title('Rainfall per day')
plt.xlabel('Rainfall in mm')
plt.ylabel('Frequency')
sns.distplot(df.Rainfall,bins=100,color='b');
```

```python
df1.Rainfall.fillna(df1.Rainfall.mode()[0],inplace = True)
```

```python
df1.drop(columns=['WindGustDir','WindDir9am','WindDir3pm'],axis=1,inplace=True)
```

```python
numerical_feature = [feature for feature in df1.columns if df1[feature].dtypes != 'O']
pop=numerical_feature.pop(2)
numerical_feature 
```

```python
for feature in numerical_feature:
    df1[feature] = df1[feature].fillna(df1[feature].mean())
```

```python
df1
```

```python
df1.RainToday.fillna(value=df1.RainToday.mode()[0],inplace = True)
df1.RainTomorrow.fillna(value=df1.RainTomorrow.mode()[0],inplace = True)
df1 = pd.get_dummies(columns=['RainToday','RainTomorrow'],data = df1,drop_first = True)
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(df1.isnull(),yticklabels = False , cbar = False ,cmap="viridis")
```

```python
df1.Date = pd.to_datetime(df1.Date)
df1.dtypes
```

```python
df1['Month'] = df1.Date.dt.month
```

```python
rainfall_per_month=df1.groupby('Month')['Rainfall'].sum()
```

```python
plt.figure(figsize=(14, 7))
plt.bar(rainfall_per_month.index,rainfall_per_month.values)
plt.xticks(labels=rainfall_per_month.index, ticks=range(1,13))
plt.xlabel("months")
plt.ylabel("rainfall in mm") #total rainfall per month from jan to dec
```

```python
df1.drop(columns=['Date','Location'],axis=1,inplace = True)
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split

x= df1.drop('RainTomorrow_Yes', axis=1)
y= df1.RainTomorrow_Yes
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2, random_state=50)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_default=DecisionTreeClassifier(max_depth=7,criterion='gini')
dt_default.fit(x_train,y_train)
y_pred1 = dt_default.predict(x_test)
print(accuracy_score(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred1))
```
